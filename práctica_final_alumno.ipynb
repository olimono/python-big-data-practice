```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

# A. Una función que calcule el precio total de la compra

def total_shop(articulos, compra):
    items_bought = map(lambda x: articulos[x], compra)
    return reduce(lambda x, y: x + y, items_bought, 0)

print(f"La compra total asciende a {total_shop(articulos, compra)}.")

# B. Una función que calcule el precio total de la compra con un descuento del 10%

def discount(articulos, compra):
    return total_shop(articulos, compra) * 0.9

print(f"La compra con descuento es {discount(articulos, compra)}.")

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def discount_if(articulos, compra):
    if total_shop(articulos, compra) > 100:
        return discount(articulos, compra)

print(f"La compra con descuento condicional es {discount_if(articulos, compra)}. Si el resultado es 139.5 es correcto, ya que la compra asciende a más de 100€")


# D. Una función que calcule el precio total aplicando el IVA (21%)

def add_iva(articulos, compra):
    return round(total_shop(articulos, compra) * 1.21, 2)

print(f"La compra con iva asciende de {total_shop(articulos, compra)} a {add_iva(articulos, compra)}")

# E.lista los artículos cuyo precio es superior a 20€

def items_over_20(articulos):
    return list(filter(lambda x: articulos[x] > 20, articulos))

print(f"Los artículos mayores de 20 euros son: {items_over_20(articulos)}.")
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca e nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]

def get_student_info(details):
    info = [details[0], details[1]]
    average = reduce(lambda x, y: x + y, details[3]) / len(details[3])
    info.append(average)
    return info

def get_students(lista, curso):
    lista_filtrada = filter(lambda x: int(x[2]) == curso, lista)
    return list(map(get_student_info, lista_filtrada))
    
print(f"Aquí la lista de alumnos de 1er curso: {get_students(alumnos, 1)}")
```
